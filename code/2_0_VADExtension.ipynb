This notebook is for VAD (Valence, Arousal and Dominance) expansion based on nltk.wordnet synonyms. The expanded word's VAD is the mean VAD of their synoymns.

For neural words or words not found in VAD_expansion, the VAD is [5,3,5].

```python
import pandas as pd
from nltk import word_tokenize, pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
from itertools import chain
from nltk.corpus import wordnet
```

```python
data = pd.read_csv('VAD.csv')
```

```python
VAD = data[['Word','V.Mean.Sum','A.Mean.Sum','D.Mean.Sum']].copy()
VAD.columns = ['Word','V','A','D']
VAD.head()
```

```python
VAD['Word'] = VAD['Word'].astype(str)
VAD.dtypes
```

```python
VAD.set_index('Word',inplace=True)
```

```python
def isVAD(x):
    return x in VAD.index
```

```python
def VADsyn(x):
    synonyms = wordnet.synsets(x)
    VAD_syn = set(chain.from_iterable([word.lemma_names() for word in synonyms]))
    inVAD = list(set(list(filter(isVAD,VAD_syn))+[x]))
    notinVAD = list(set(VAD_syn) - set(inVAD))
    print(notinVAD,'\n')
    
    for syn in notinVAD:
        if syn not in dict_syn.keys():
            dict_syn.update({syn:inVAD})
        else:
            dict_syn.update({syn:list(set(dict_syn[syn]+inVAD))})
```

```python
dict_syn = {}

for word in VAD.index:
    print(word,type(word))
    VADsyn(word)
```

```python
VAD_extend = pd.DataFrame(columns=['Word','V','A','D'])
```

```python
VAD_extend = pd.DataFrame(columns=['Word','V','A','D'])
for word in dict_syn.keys():
    print(word)
    name = [word]
    value = list(VAD.loc[dict_syn[word]].mean().get_values())
    VAD_extend = VAD_extend.append(pd.Series(name+value,index=VAD_extend.columns),ignore_index=True)
```

```python
VAD_extend.set_index('Word',inplace=True)
VAD_extend_all = VAD.append(VAD_extend)
```

```python
VAD_extend_all.to_csv('VAD_extend.csv',index=False)
```

```python
VAD_extend_all.reset_index(inplace=True)
```

```python
VAD_dict = VAD_extend_all.to_dict('index')
```
